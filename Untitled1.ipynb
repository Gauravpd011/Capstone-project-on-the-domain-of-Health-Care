```python
def bucket_year(classes):
    
    if classes <= 60:
        return 'Low'
    elif classes >61 and classes <=120:
        return 'medium'
    elif classes > 121 and classes <=240:
        return 'High'
    else:
        return 'Very High'
```

```python
df['iralcfy']=df['iralcfy'].apply(bucket_year)
df['irmjfy']=df['irmjfy'].apply(bucket_year)
df['ircocfy']=df['ircocfy'].apply(bucket_year)
df['irhallucyfq']=df['irhallucyfq'].apply(bucket_year)
df['irinhalyfq']=df['irinhalyfq'].apply(bucket_year)
df['irherfy']=df['irherfy'].apply(bucket_year)
df['irmethamyfq']=df['irmethamyfq'].apply(bucket_year)
df['cig30use']=df['cig30use'].apply(bucket_year)
```

```python
def bucket_hour(hours):
    
    if hours <= 15:
        return 'Low'
    elif hours >16 and hours <=30:
        return 'medium'
    elif hours > 31 and hours <=45:
        return 'High'
    else:
        return 'Very High'
```

```python
df['wrkdhrswk2']=df['wrkdhrswk2'].apply(bucket_hour)
```
